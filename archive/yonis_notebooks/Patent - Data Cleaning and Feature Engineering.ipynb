```python
#cleaning function packages
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
stop = stopwords.words('english')
wl = WordNetLemmatizer()
```

```python
#clean data function
def clean_data(text):
  word_tokens = word_tokenize(text) #split into words
  clean_words = []
  for word  in word_tokens:
    if word.isalpha() and word.lower() not in stop:
        clean_word = wl.lemmatize(word.lower())
        clean_words.append(clean_word)
  clean_text = " ".join(clean_words)
  return clean_text
```

```python
#applying cleaning function
df['clean_abstract'] = df['abstract'].apply(clean_data)
```

**Feature Engineering:**

```python
#importing packages
from scipy import  stats
!pip install https://github.com/andreasvc/readability/tarball/master
import readability
!pip install syntok
import syntok.segmenter as segmenter
```

```python
from statistics import mean,median
```

```python
#functions
#word median
def med_word_length(para):
  word_list = []
  for word in para.split():
    word_list.append(len(word))
  return median(word_list)

#avg word length
def avg_word_length(para):
  word_list = []
  for word in para.split():
    word_list.append(len(word))
  return mean(word_list)

#skewness
def skew_word_length(para):
  word_list = []
  for word in para.split():
    word_list.append(len(word))
  return stats.skew(word_list)

#get kinkade_score
def get_kinkade(abstract):
  tokenized = '\n\n'.join('\n'.join(' '.join(token.value for token in sentence) for sentence in paragraph) for paragraph in segmenter.analyze(abstract))
  results = readability.getmeasures(tokenized, lang='en')
  grade = results['readability grades']['Kincaid']
  return grade

```

```python
df['kinkade'] = df['abstract'].apply(get_kinkade)
```

```python
df['median_word_length'] = df['abstract'].apply(med_word_length)
```

```python
df['avg_word_length'] = df['abstract'].apply(avg_word_length)
```

```python
df['skew_word_length'] = df['abstract'].apply(skew_word_length)
```
